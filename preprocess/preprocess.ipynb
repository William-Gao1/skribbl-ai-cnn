```python
from get_image import get_img_from_strokes

import os
import numpy as np
import tqdm
import ndjson

from multiprocessing import Pool

nthreads = 8
target_shape = (64, 64)

scratch = os.path.join(os.path.expanduser('~'), "scratch")
loc = os.path.join(scratch, "small-quickdraw")

filenames = os.listdir(loc)

dest_dir = os.path.join(scratch, f"{target_shape[0]}-pre-quickdraw/")
os.makedirs(dest_dir)
```

```python
def process_file(file, target_shape=(64,64)):
  source = os.path.join(loc, file)
  dest = os.path.join(dest_dir, file.replace(".ndjson", ".npy"))
  
  with open(source, 'r') as f:
    data = ndjson.load(f)
  
  result = np.empty((len(data), target_shape[0], target_shape[1]))
  
  for i in tqdm.tqdm(range(len(data))):
    bitmap = get_img_from_strokes(data[i]["drawing"], target_shape)
    result[i] = bitmap
  
  np.save(dest, result)
```

```python
with Pool(nthreads) as p:
  p.map(process_file, filenames)
```
